````python
def filtered_pokedex(new_pokedex, filter_options):
    matches  = []
    for pokemon_id, values  in new_pokedex.items():
        match = True 
        
        for category, value in filter_options.items():
            category_match = False 
            
            category_type = type(value)
            if category_type is str:
                if new_pokedex[pokemon_id][category].lower() == value.lower():
                    category_match = True 
                
            elif category_type is float or category_type is int:
                if new_pokedex[pokemon_id][category] >= value:
                    category_match = True 
                
            if category_match and match:
                    match= True 
                    
            else: 
                    match = False
        if match:
            matches.append(values)
            

    return matches 
            
    

   
    
filter_options = {
   'Attack':   87,
   'Defense':  79,
    'Type':     'Electric'
}
    
filtered_pokedex(new_pokedex, filter_options)    
````
