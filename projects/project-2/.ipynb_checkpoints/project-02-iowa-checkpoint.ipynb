```python
iowa
```

```python
county_growth_sales = pd.pivot_table(iowa, values=['Sale (Dollars)'], index=['County'], columns=['year',quarter], aggfunc=np.sum)
```

```python
n_rows= county_sales.shape[0]

q1_16_ix = -1
q1_15_ix = 1
```

```python
county_growth_sales_growth = [county_growth_sales.iloc[row_i][q1_16_ix]/county_growth_sales.iloc[row_i][q1_15_ix] for row_i in range(0, n_rows)]
```

```python
county_growth_sales[('growth')]= county_growth_sales_growth
```
