### 6.1 Create columns for Q1 2015 sales and Q1 2016 sales.

The owner is not just interested in what locations are currently selling well, but also which areas are showing the most promising sales growth from 2015 to 2016. 

For yearly changes we will focus on the changes from Q1 2015 to Q1 2016 specifically (avoiding effects of season changes).

Note: Q1 is months: January, February, and March.


For this analysis I am going to go back to my original Iowa Data Frame because the open_15 data frame is losing key values. If a county has many stores that closed down in the past year, its likely the market is not growing in overall sales. if we exclude stores that shut down in this analysis, we may think profits are growing(other stores in the county take some marginal volume of the sales from the shut down stores) when they are actually shrinking. Next I am going to identify two key factors: Q1 Growth and Q1 profits / Gal. these will show not only whether the county has a high population/growth, but also whether or not there are residents with expensive tastes.

```python
county_growth_sales = pd.pivot_table(iowa, values=['Sale (Dollars)'], index=['County'], columns=['year',quarter], aggfunc=np.sum)
```

```python
n_rows = county_growth_sales.shape[0]
q1_16_ix = -1
q1_15_ix = 1
```

```python
#n_rows = qtable.shape[0]
#year_2015_index = 2
#year_2016_index = -2
#for row_i in range(0,n_rows):
    #print qtable.iloc[row_i][year_2016_index]/qtable.iloc[row_i][year_2015_index]
    #break
```

```python
county_growth_sales_growth = [county_growth_sales.iloc[row_i][q1_16_ix]/county_growth_sales.iloc[row_i][q1_15_ix] for row_i in range(0, n_rows)]
```

```python
county_growth_sales[('growth')]= county_growth_sales_growth
```

```python
county_growth_sales['sum_total'] = county_growth_sales.iloc[:,0:5].sum(axis = 1)
```

```python
county_growth_sales
```

### 8.2 Calculate maximum loan interest rate.

The owner unfortunately has to take out a loan to to pay for the entirety of the property and liquor. 

**Conditions of the loan**:

Every quarter, (except for Q1 when he takes out the loan), the interest rate is applied to the remaining balance on the loan and added to his debt. 

If, for example, he made no payments whatsoever, not only would the amount of debt he owed keep increasing by quarter but the debt incurred from interest would keep increasing as well. (Since his debt keeps growing, the interest rate is applied each month to the bigger balance, incurring more debt each time.)

**Determining maximum acceptable interest:**

Given this, _the owner wants to be 99% confident that he will be able to pay off his loans._ If the interest rate is too high he will have to eventually declare bankruptcy.

Based on his expected profits per quarter (assume the expected profit stays the same for all future quarters), what is the maximum interest rate he should accept on the loan to be 99% confident he will be able to pay it off?
