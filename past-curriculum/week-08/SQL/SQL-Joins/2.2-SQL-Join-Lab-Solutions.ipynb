### 1. How many products per category does the catalog contain?
Print the answer with the CategoryName, and Count.

```python
SQL_STRING = '''

SELECT C."CategoryName", count(*) AS "Category Count"
FROM Products AS P
INNER JOIN Categories AS C
ON C."CategoryID" = P."CategoryID"
GROUP BY C."CategoryName"
ORDER BY 2 DESC

'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

### 2. What 5 customers are generating the highest revenue?
Print a table with CustomerID and Total Revenue. You will need to use data from 3 tables.

```python
SQL_STRING = '''

SELECT DISTINCT "Discount"
From Order_Details OD
LIMIT 5
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

```python
SQL_STRING = '''

SELECT C."CustomerID",C."CompanyName", SUM(OD."UnitPrice" * (1-OD."Discount") * OD."Quantity") AS "Revenue"
FROM Orders O
INNER JOIN Order_Details OD
ON O."OrderID" = OD."OrderID"
INNER JOIN Customers C
ON C."CustomerID" = O."CustomerID"
GROUP BY C."CustomerID", C."CompanyName"
ORDER BY 3 DESC
LIMIT 5
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

### 3. In which country are the top 5 suppliers by number of units supplied?
Print a table with the supplier's CompanyName, Country and total number of units supplied.

```python
SQL_STRING = '''

SELECT S."CompanyName", S."Country", SUM(P."UnitsInStock") AS "Units Stocked"
FROM Suppliers S
INNER JOIN Products P
on S."SupplierID" = P."SupplierID"
GROUP BY S."CompanyName", S."Country"
ORDER BY 3 DESC
LIMIT 5
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

### 4. What are the terrritories with the greatests revenue. 

Return TerritoryDescription and Revenue. 

```python
SQL_STRING = '''

SELECT T."TerritoryDescription", SUM(OD."UnitPrice" * (1-OD."Discount") * OD."Quantity") AS "Revenue"
FROM Orders O
INNER JOIN Order_Details OD
ON OD."OrderID" = O."OrderID"
INNER JOIN Employees E
ON E."EmployeeID" = O."EmployeeID"
INNER JOIN EmployeeTerritories ET 
ON ET."EmployeeID" = E."EmployeeID"
INNER JOIN Territories T
ON T."TerritoryID" = ET."TerritoryID"
GROUP BY T."TerritoryDescription"
--HAVING SUM(OD."UnitPrice" * (1-OD."Discount") * OD."Quantity") > 166537
ORDER BY 2 DESC, 1 ASC
LIMIT 10
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

#### 5) For each product print out the ProductID and the count of items that are priced less than that item. 

```python
SQL_STRING = '''

SELECT *
FROM (SELECT P."ProductID",P."UnitPrice"
FROM Products P
ORDER BY P."UnitPrice" ASC
LIMIT 5) SUB1
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```

```python
SQL_STRING = '''

SELECT SUB1."ProductID", SUB1."UnitPrice", count(*)
FROM (SELECT P."ProductID",P."UnitPrice"
FROM Products P
ORDER BY P."UnitPrice" ASC
LIMIT 5) SUB1

LEFT JOIN
(SELECT P."ProductID",P."UnitPrice"
FROM Products P
ORDER BY P."UnitPrice" ASC
LIMIT 5) SUB2

ON SUB1."UnitPrice" > SUB2."UnitPrice"
--WHERE SUB1."UnitPrice" > SUB2."UnitPrice"
GROUP BY SUB1."ProductID", SUB1."UnitPrice"
ORDER BY SUB1."UnitPrice" ASC
'''

df = pd.read_sql(SQL_STRING, con=conn)
df
```
