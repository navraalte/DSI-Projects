```python
import pandas as pd
import numpy as np
```

```python
hsq_path = '~/DSI-SF-5/datasets/humor_styles/hsq_data.csv'
df = pd.read_csv(hsq_path)
```

```python
df.head()
```

```python
def get_mean(series):
    return np.mean(series)
```

```python
df[["age", "agressive", 'selfdefeating']].apply(get_mean, axis=0) # axis=0 --> operate on cols
```

```python
# access a subset of the rows
# df.loc[row_index]
df.loc[0:2]
```

```python
# access a subset of rows and cols
# df.loc[row_Index, col_index]
df.loc[0:2, "accuracy"]
```

```python
# access a subset of rows and cols
# df.loc[row_Index, col_index]
df.loc[0:2, ["accuracy", "gender"]]
```

```python
# returns all indicies specified 0,1,2,3 totally to 4 rows
df.loc[0:3]
```

```python
# only accepts a row_index argument and not col_index
# returns 0:n-1 indicies specified 0,1,2 totally to 3 rows 
df.iloc[0:3]
```
