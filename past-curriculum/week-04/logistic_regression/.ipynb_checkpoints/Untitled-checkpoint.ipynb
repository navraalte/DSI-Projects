```python
def loss(prob, label):
    prob_pos = prob[np.where(label == 1)]
    prob_neg = 1 - prob[np.where(label == 0)]
    return -np.mean(np.log(np.hstack((prob_pos, prob_neg))))
np.exp(-loss(prob[:, 1], y))
```
